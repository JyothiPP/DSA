```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#To read the dataset
data=pd.read_excel("/content/iris.xls")
data
```

```python
#To display the columns in the dataset
data.columns
```

```python
#To calculate the mean of each column
data[['SL','SW','PL','PW']].mean()
```

```python
#To check for the null values
data.info()
```

VISUALIZATIONS USING THE DATASET

```python
#To plot a histogram that shows the distribution of Sepal Length of Iris
plt.hist(data['SL'],rwidth=0.6)
plt.title('DISTRIBUTION OF SEPAL LENGTH')
plt.show()
```

```python
#To plot a bar chart for comparing the Petal Widths of the three species of Iris
plt.bar(data['Classification'],data['PW'])
plt.xlabel('Classification',c='green')
plt.ylabel('Petal Width',c='green')
plt.title('COMPARISON OF PETAL WIDTHS')
plt.show()
```

```python
#To compare the distribution of Sepal Width and Petal Width of Iris-setosa using KDE
data_1=data.groupby('Classification').get_group('Iris-setosa')
data_1[['SW','PW']].plot.kde()
```

```python
#To create a scatter plot using Sepal Length and Petal Length of the three species of Iris
data_2=data.groupby('Classification').get_group('Iris-versicolor')
data_3=data.groupby('Classification').get_group('Iris-virginica')
plt.scatter(data_1['SL'],data_1['PL'],color='red')
plt.scatter(data_2['SL'],data_2['PL'],color='green')
plt.scatter(data_3['SL'],data_3['PL'],color='blue')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.show()
```
